```python
#import modules and read dataset from file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 100)
#pd.set_option('display.max_rows', 300)

df_cellPhones = pd.read_csv('cell_phones_total.csv')

#fill in missing values (replaces all NaN, None or pd.NA) 
df_cellPhones.fillna(value=pd.NA, inplace=True)

# Remove columns with all missing values 
df_cellPhones.dropna(axis="columns", how="all", inplace=True)

# Remove rows with all missing values
df_cellPhones.dropna(axis="rows", how="all", inplace=True)

# Remove columns with all zeros
# -- Implementation to be defined --

# Remove rows with all zeros
# -- Implementation to be defined --

# Remove columns up to 1979, since no cellphones by then
#df_cellPhones = df_cellPhones.loc[: , df_cellPhones.columns > "1979"]


# Define a function that converts cell values to a float or NaN
def convert_to_number(n):
    
    if isinstance(n, str):
        
        n = n.strip()
            
        if n.endswith('k'):
            return float(n[:-1]) * 1e3
        
        elif n.endswith('M'):
            return float(n[:-1]) * 1e6
        
        elif n.endswith('B'):
             return float(n[:-1]) * 1e9
        
        elif n == '':
            return np.nan
        else:
            float(n)
    elif pd.isnull(n):
        return np.nan
    else:
        n

        
#select columns representig years
year_columns = df_cellPhones.columns.drop('iso-3')

#loop through columns and convert entries to float
for col in year_columns:
    df_cellPhones[col] = df_cellPhones[col].apply(convert_to_number)
    
# Make sure all data is numeric
df_cellPhones[year_columns] = df_cellPhones[year_columns].apply(pd.to_numeric, errors='coerce')

# Replace NaN values with zeros until the first non-NaN value in a row.
# NaN values between non-NaN values are replaced by previous non-NaN value
df_cellPhones.iloc[:, 1:] = df_cellPhones.iloc[:, 1:].apply(lambda row: row.ffill().fillna(0), axis=1)

# Select columns to display
selected_columns = ['iso-3', '2015', '2016', '2017', '2018', '2019']

# Display table for selected columns
df_cellPhones[selected_columns].head()   
```



### 1.2 Cleaning dataset

Read dataset about the population size of countries over years.
Prepare and clean the dataset.


**Code:**

```python
df_pop = pd.read_csv("pop_data.csv")

# Make sure to be able to display all columns and rows
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 300)

# Display part of data
df_pop.head()

# Check the datatype of columns to make sure it is float object
#df_pop.info()

#fill in missing values (replaces all NaN, None or pd.NA) 
df_pop.fillna(value=pd.NA, inplace=True)

# Check if there are entries containing empty space
num_empty = df_pop.map(lambda x: isinstance(x, str) and x.strip() == "").sum().sum()
print(f"Number of empty strings or spaces: {num_empty}")

# Remove first column in dataset due to irrelevant data
df_pop.drop(df_pop.columns[0], axis=1, inplace=True)

# Select all columns representing years
columns_to_check = df_pop.loc[:, "1960":].columns

# Check rows with all missing values
rows_with_all_nan = df_pop[columns_to_check].isna().all(axis=1)

# Remove rows with all missing values
df_pop.drop(index=df_pop[rows_with_all_nan].index, inplace=True)

# Check rows with any missing values
rows_with_any_nan = df_pop[df_pop.isna().any(axis=1)]

# Display table 
df_pop.head()

```

### 1.3 Reshape datasets

To combine the datasets for cell phones and population, the datasets need to be pivoted from
wide to long format first.



```python


# Pivot dataframe population to long format
long_df_pop = pd.melt(df_pop, id_vars="iso-3", var_name="year", value_name="population")


# Express the population size with 6 significant numbers
long_df_pop["population"] = long_df_pop["population"].apply(lambda x: f"{x:.6g}")

# After join, values become of object type. Need to convert to float
long_df_pop["population"] = pd.to_numeric(long_df_pop["population"], errors="coerce")

# Display table
long_df_pop.head()



```

```python


# Pivot dataframe cellphones to long format
long_df_cellPhones = pd.melt(df_cellPhones, id_vars="iso-3", \
                             var_name="year", value_name="n_cellphones")

# Express the population size with 6 significant numbers
long_df_cellPhones["n_cellphones"] = long_df_cellPhones["n_cellphones"].apply(lambda x: f"{x:.6g}")

# After join, values become of object type. Need to convert to float
long_df_cellPhones["n_cellphones"] = pd.to_numeric(long_df_cellPhones["n_cellphones"], errors="coerce")


# Display table
long_df_cellPhones.head()

```

### 1.4 Joining datasets

The two datasets need to be inner joined, i.e. keeping the intersection part


```python

# Merge the two datasets , inner join
df_cellphone_pop = pd.merge(long_df_cellPhones, long_df_pop)

# Display table
df_cellphone_pop.head()


```

##

Read country data to map country code to country name


```python

# Read data
df_country_code = pd.read_csv("country_data.csv")

#display content
df_country_code.head()

```

```python


# Select relevant columns for country name and country code
subset_df_country_code = df_country_code[["name", "alpha-3"]]

# Dispay table
subset_df_country_code.head()
```

## 

Inner join the dataset for cellphone population and country name 



```python

# Merge two datasets using corresponding keys
merged_data = pd.merge(df_cellphone_pop, subset_df_country_code,\
                       left_on="iso-3", right_on="alpha-3", how="inner")

# Dispay table
merged_data.head()

```

## 

Remove columns that consist name codes


```python
merged_data.drop(columns=["iso-3", "alpha-3"], inplace=True)

merged_data.head()
```

### 1.5 Calculate data and update dataset

Calculate the number of cellphones per capita and store the date in a new column.



```python

# Create a column to store derived data, i.e. cellphones per capita
# If population is missing, fill in with Nan instead
merged_data["cellph_per_c"] = merged_data.apply(lambda row: row["n_cellphones"]/row["population"] \
                                                if pd.notna(row["population"]) else np.nan, axis=1)
    
merged_data.head()   

```

### 1.6 Reshape final dataset

The merged data need to be converted to wide format



```python

# Convert to wide format
wide_merged_data = merged_data.pivot(index="year", columns="name", values="cellph_per_c")

# Sort columns by value in 2019 in descending order 
wide_merged_data.sort_values(by="2019", axis=1, ascending=False, inplace=True)

# Selet columns to display
selected_columns = wide_merged_data.columns[:5]

#wide_merged_data.columns.name = "year"

# Display table
wide_merged_data.tail()[selected_columns]

```
