## Part 1 - Joining data

Combine cellphone data with population data to analyse the number of 
cell phones per capita.


Prepare and clean the dataset, cell_phones, from pevious assignment


**Code:**


```python
#import modules and read dataset from file
import pandas as pd
import numpy as np

df_cellPhones = pd.read_csv('cell_phones_total.csv')

#fill in missing values (replaces all NaN, None or pd.NA) 
df_cellPhones.fillna(value=pd.NA, inplace=True)

# Remove columns with all missing values 
df_cellPhones.dropna(axis="columns", how="all")


#convert cell values to a float or NaN
def convert_to_number(n):
    
    if isinstance(n, str):
        
        n = n.strip()
            
        if n.endswith('k'):
            return float(n[:-1]) * 1e3
        
        elif n.endswith('M'):
            return float(n[:-1]) * 1e6
        
        elif n.endswith('B'):
             return float(n[:-1]) * 1e9
        
        elif n == '':
            return np.nan
        else:
            float(n)
    elif pd.isnull(n):
        return np.nan
    else:
        n

#select columns representig years
year_columns = df_cellPhones.columns.drop('iso-3')

#loop through columns and convert entries to float
for col in year_columns:
    df_cellPhones[col] = df_cellPhones[col].apply(convert_to_number)
    

# Make sure all data is numeric
df_cellPhones[year_columns] = df_cellPhones[year_columns].apply(pd.to_numeric, errors='coerce')

# Select columns to display
selected_columns = ['iso-3', '2015', '2016', '2017', '2018', '2019']

# Select year columns 1960 - 1980 where no available cellphones
columns_to_fill = [col for col in df_cellPhones.columns if col.isdigit() and int(col)<= 1980]

#fill all NaN entries with 0 for the selected columns 
df_cellPhones[columns_to_fill] = df_cellPhones[columns_to_fill].fillna(0)

# Display table for selected columns
df_cellPhones[selected_columns].head()   
```


Read dataset about the population size of countries over years.
Prepare and clean the dataset.


**Code:**

```python
df_pop = pd.read_csv("pop_data.csv")

# Make sure to display all columns
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 300)

# Display part of data
df_pop.head()

# Check the datatype of columns to make sure it is float object
#df_pop.info()

#fill in missing values (replaces all NaN, None or pd.NA) 
df_pop.fillna(value=pd.NA, inplace=True)

# Check if there are entries containing empty space
num_empty = df_pop.map(lambda x: isinstance(x, str) and x.strip() == "").sum().sum()
print(f"Number of empty strings or spaces: {num_empty}")

# Remove first column in dataset due to irrelevant data
df_pop.drop(df_pop.columns[0], axis=1, inplace=True)

# Select all year-columns
columns_to_check = df_pop.loc[:, "1960":].columns

# Check rows with all missing values
rows_with_all_nan = df_pop[columns_to_check].isna().all(axis=1)

# Remove rows with all missing values
df_pop.drop(index=df_pop[rows_with_all_nan].index, inplace=True)

# Display table
df_pop.head()

```



To combine the datasets for cell phones and population, the datasets need to be pivoted from
wide to long format first.



```python
# Pivot dataframe population to long format
melt_df_pop = pd.melt(df_pop, id_vars="iso-3", var_name="year", value_name="population")

# Express the population size with 6 significant numbers
melt_df_pop["population"] = melt_df_pop["population"].apply(lambda x: f"{x:.6g}")

# Display table
melt_df_pop.head()
```

```python


# Pivot dataframe cellphones to long format
melt_df_cellPhones = pd.melt(df_cellPhones, id_vars="iso-3", \
                             var_name="year", value_name="n_cellphones")
# Display table
melt_df_cellPhones.head()

```

#

The two datasets need to be joined, keeping the intersection part


```python
# Merge the two data sets , inner join
df_cellphone_pop = pd.merge(melt_df_cellPhones, melt_df_pop)

# Display table
df_cellphone_pop.head()

```
